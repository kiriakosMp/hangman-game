```python
import random
from IPython.display import clear_output
from time import sleep
import os
```

```python
gallow=[[' ','_','_','_','_','_','_','_','_','_','_','_','_','_',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ','|',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' '],
       [' ','|',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' '],
       ['_','|','_',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']]
```

```python
class Game:
    
    body={0:[2,-2,'o'],1:[3,-2,'|'],2:[3,-3,'/'],3:[3,-1,'\ '],4:[4,-3,'/'],5:[4,-1,'\ ']}
    wrong_letters=set()
    given_letter=''
    underscores=[]
    counter=0
        
    
    def __init__(self,word,gallow):
        self.word=word.lower()
        self.gallow=gallow
        
    def welcome_the_player(self,usr):
        print(f'Welcome {usr} to the Hangman game!')
        sleep(1)
        print("Let's play!")
        sleep(1)
                
    def print_the_gallow(self):
        for i in self.gallow:
            print(''.join(i))
            
    def hide_the_word(self):
        for i,char in enumerate(self.word):
            if char==' ':
                self.underscores.append(' ')
                continue
            self.underscores.append('_')
        print('Word: ',' '.join(self.underscores))
    
    def ask_for_letter(self):
        while True:
            letter=input('Give a letter: ')
            if letter.isalpha() and len(letter)==1:
                break
            print(f'{letter} is not an acceptable answer!')
            sleep(1)
        letter=letter.lower()
        self.given_letter=letter
        
    def check_conditions(self):
        
        if self.given_letter in self.underscores or self.given_letter.upper() in self.underscores:
            print('You have already chosen this letter and it was correct!')
            sleep(2)

        if self.given_letter in self.wrong_letters:
            print("Don't mention this letter again!")
            sleep(1)

        if self.given_letter in self.word:
            indices=[i for i,let in enumerate(self.word) if let==self.given_letter]
            
#___________________________________________________If it exists, it is printed in its original form    
            for i in indices:
                if i==0 or self.word[i-1]==' ':
                    self.underscores[i]=self.given_letter.upper()
                    continue
                self.underscores[i]=self.given_letter
                
#_____________________________________If it is not exists, counter+=1 and a part of the body appears in the gallow     
        if self.given_letter not in self.word and self.given_letter not in self.wrong_letters:
            self.counter+=1
            if self.counter ==1:
                self.gallow[self.body[0][0]][self.body[0][1]]=self.body[0][2]
            self.gallow[self.body[self.counter][0]][self.body[self.counter][1]]=self.body[self.counter][2]
            self.wrong_letters.add(self.given_letter)
            
            
    def check_if_wins(self):
        if not '_' in self.underscores:
            print('\nWin!')
            return True
        return False
    
    def check_if_loses(self,word):
        if self.counter==5:
            print("\nThe end! Our guy didn't survive. :(")
            print(f'The word was : {word}')
            sleep(2)
            return True
        return False
    
    def clear(self):
        spaces={0:[2,-2,' '],1:[3,-2,' '],2:[3,-3,' '],3:[3,-1,'  '],4:[4,-3,' '],5:[4,-1,'  ']}
        for i in spaces:
            self.gallow[spaces[i][0]][spaces[i][1]]=spaces[i][2]
        self.underscores=[]
        self.wrong_letters=set()
        self.counter=0
        
    def ask_to_play_again(self):
        while True:
            again = input('Play again? (y/n)')
            if again[0].lower()=='y':
                return True
            elif again[0].lower()=='n':
                return False
            else:
                print('Please type y to play again or n to stop!')
                continue
                
    def __str__(self):
        print('The word is',self.word)
        return ''
```

```python
class Difficulty:
    
    def __init__(self):
        self.path = os.getcwd()
        self.dict_of_dif={'easy':[],'medium':[],'hard':[]}

    
    
    
    
    def create_path(self,category_choice):
        self.path=os.getcwd()+'\\lists\\'+category_choice+'.txt'
        
    def show_list(self):
        with open(self.path,'r',encoding='utf-8-sig') as f:
            print(f.read())
    
    def dict_of_dif_levels(self):
        with open(self.path,'r',encoding='utf-8-sig') as f:
            all_text = f.read()
            words = list(map(str, all_text.split('\n')))
            self.dict_of_dif={'easy':[],'medium':[],'hard':[]}
            for i in words:
                if 4<=len(set(i))<7:
                    self.dict_of_dif['easy'].append(i)

                elif 7<=len(set(i))<=8:
                    self.dict_of_dif['medium'].append(i)

                elif len(set(i))>8:
                    self.dict_of_dif['hard'].append(i)
            #print(self.dict_of_dif)
    
    def choose_random_word(self,player_dif_choice,player_archive):
        random_word=random.choice(list(set(self.dict_of_dif[player_dif_choice])-set(player_archive)))
        random_word=str(random_word)
        #print(random_word,len(random_word))
        return random_word
```

```python
class Player:
    
    path=os.getcwd()+"\\players' scores\\"
    
    
    def __init__(self,username):
        self.username=username
        self.archive=[]
        try:
            with open(self.path+f'{self.username}_score.txt','r') as f:
                self.total_score=float(f.read())
        except FileNotFoundError:
            self.total_score=0
            
    def ask_for_category(self):
        categories=['countries','capitals','us_states','top_movies']
        for i,category in enumerate(categories):
            print('Type',i,'for',category)
            
        choice='smth'
        within_range = False
        
        while not choice.isdigit() or not within_range:
            choice=input('\nChoose a list(0-3): ')
            if not choice.isdigit():
                print("\nSorry, I need a digit!")
                
            if choice.isdigit():
                if int(choice) in range(0,4):
                    within_range = True
                else:
                    within_range = False
                    print('\nA digit between 0 and 3 please!')
        
        return categories[int(choice)]
            
    def choose_difficulty(self):
        while True:
            difficulty=input('Choose difficulty(Easy,Medium,Hard): ')
            if difficulty[0].lower() in ['e','m','h']:
                break
        if difficulty[0].lower() == 'e':
            return 'easy'
        elif difficulty[0].lower() == 'm':
            return 'medium'
        else:
            return 'hard'
    
    def difficulty_to_points(self,diff_to_points):
        if diff_to_points=='easy':
            return 2
        elif diff_to_points=='medium':
            return 5
        else:
            return 10
        
    def set_score(self,round_score):
        self.total_score+=round_score
        with open(self.path+f'{self.username}_score.txt','w') as f:
            f.write(str(self.total_score))
            
    def show_score(self):
        try:
            with open(self.path+f'{self.username}_score.txt','r') as f:
                print('Score is',f.read())
        except FileNotFoundError:
            print('Score is 0')
            
    def archive_with_already_used_phrases(self,phrase):
        self.archive.append(phrase)
```

```python
def main():
    current_user=input("\n\nWho's playing? ")
    print('\n\n')
    player=Player(current_user)
    print(f'Welcome {current_user} to the Hangman game!\n\n')
    sleep(1)
    print("Let's play!\n\n")
    sleep(1)
    while True:
        clear_output()
        #game.clear()
        category=player.ask_for_category()
        diff=player.choose_difficulty()
        word_difficulty=Difficulty()
        word_difficulty.create_path(category)
        word_difficulty.dict_of_dif_levels()
        if not set(word_difficulty.dict_of_dif[diff])-set(player.archive):
            print('No more words/phrases in this category and dif level left! Choose another category or level, please!')
            sleep(2)
            continue
        rnd_word=word_difficulty.choose_random_word(diff,player.archive)

        game=Game(rnd_word,gallow)

        clear_output()
        game.clear()
        print('Category:',category)
        print('Difficulty:',diff)
        game.print_the_gallow()
        game.hide_the_word()
        #print(game)

        status = False
        while True:
            game.ask_for_letter()
            game.check_conditions()
            clear_output()
            print('Category:',category)
            print('Difficulty:',diff)
            game.print_the_gallow()
            print(' '.join(game.underscores))
            print('\nWrong guesses: ',[i for i in game.wrong_letters])
            if game.check_if_wins():
                points=player.difficulty_to_points(diff)
                player.set_score(points)
                sleep(2)
                clear_output()
                print('\n\n\nGet ready for the next round!\n\n\n')
                sleep(3)
                status = True
                break
            elif game.check_if_loses(rnd_word):
                points=player.difficulty_to_points(diff)
                points=-(1/points)
                player.set_score(points)
                break
        player.archive_with_already_used_phrases(rnd_word)
        player.show_score()
        if status:
            continue
        if not game.ask_to_play_again():
            break
```

```python
main()
```
